## Deaths of Shelter Residents in Toronto From 2007-2024 

* The "Annual Review of Statistics on Deaths of Shelter Residents" highlights some key points about the ongoing crisis affecting people experiencing homelessness in Toronto.
* The data shows a concerning trend where the leading cause of death among homeless individuals is drug toxicity, accounting for nearly half of the recorded deaths in 2022.
* This is reflective of a broader drug toxicity crisis in Toronto, where potent and unregulated drugs are increasingly contributing to fatal overdoses.

* Additionally, the median age of death for homeless individuals remains alarmingly low compared to the general population—52 for men and 48 for women in 2023.
* This disparity underscores the severe health risks and shorter lifespans associated with homelessness. Other causes of death include cardiovascular disease and cancer, but many deaths still fall into the "unknown or pending" category, showing gaps in available data.

* Toronto's public health agencies continue to work on harm reduction efforts, including providing safer drug supplies and treatment programs, to combat the crisis.
* However, long-term solutions like affordable housing and mental health services are critical to improving the overall health outcomes for this vulnerable population​

## Data Cleaning and Preprocessing

```python
# looking for duplicates
df = df.drop_duplicates()
```

```python
# looking for null values
df.isnull().sum()
```

```python
missing_data_percentage = df['Transgender/Non-binary/Two-Spirit'].isna().mean() * 100
print(f"Percentage of missing data in 'Transgender/Non-binary/Two-Spirit': {missing_data_percentage}%")
```

```python
#imputating missing values
df['Transgender/Non-binary/Two-Spirit'] = df['Transgender/Non-binary/Two-Spirit'].fillna(0)
df['Transgender/Non-binary/Two-Spirit']
```

```python
total_deaths = df[['Male', 'Female', 'Transgender/Non-binary/Two-Spirit']].sum()
total_deaths
```

## Correlation

```python
correlation_matrix = df[['Total decedents', 'Male', 'Female', 'Transgender/Non-binary/Two-Spirit']].corr()
print(correlation_matrix)

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

# Visualizations

```python
# Plot each category
plt.plot(df['Year'], df['Male'], marker='o', label='Male')
plt.plot(df['Year'], df['Female'], marker='o', label='Female')
plt.plot(df['Year'], df['Transgender/Non-binary/Two-Spirit'], marker='o', label='Transgender/Non-binary/Two-Spirit')

# Add titles and labels
plt.title('Yearly Representation by Gender')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
# Sum the values for each category across all years
total_male = df['Male'].sum()
total_female = df['Female'].sum()
total_other = df['Transgender/Non-binary/Two-Spirit'].sum()

# Prepare data for the pie chart
labels = ['Male', 'Female', 'Transgender/Non-binary/Two-Spirit']
sizes = [total_male, total_female, total_other]

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#66b3ff', '#ff9999', '#99ff99'])

# Equal aspect ratio ensures that the pie is drawn as a circle
plt.axis('equal')

# Add title
plt.title('Gender Representation Across All Years')

# Show the plot
plt.show()
```

```python
df.groupby('Year')[['Male', 'Female', 'Transgender/Non-binary/Two-Spirit']].sum().plot(kind='bar', stacked=True)

plt.title('Yearly Deaths Distribution by Gender')
plt.xlabel('Year')
plt.ylabel('Number of Decedents')
plt.show()
```

```python
df[['Male', 'Female', 'Transgender/Non-binary/Two-Spirit']].plot(kind='box')

plt.title('Box Plot of Deaths by Gender')
plt.ylabel('Number of Decedents')
plt.show()
```

```python
df.groupby('Month')['Total decedents'].sum().plot(kind='bar', color='purple')

plt.title('Total Deaths per Month')
plt.xlabel('Month')
plt.ylabel('Number of Decedents')
plt.show()
```

```python
df.groupby('Month')[['Male', 'Female', 'Transgender/Non-binary/Two-Spirit']].sum().sort_values(by='Month', ascending=False).plot(kind='line', marker='o')

plt.title('Monthly Deaths Trend by Gender')
plt.xlabel('Month')
plt.ylabel('Number of Decedents')
plt.show()
```

```python
df.set_index('Year', inplace=True)
df['Total decedents'].rolling(window=3).mean().plot()
plt.title('Rolling Average of Total Decedents Over Years')
plt.show()
```

```python
# Create a new column for total deaths across all genders
df['Total'] = df['Male'] + df['Female'] + df['Transgender/Non-binary/Two-Spirit']

# Plot a histogram of the total deaths across all years
plt.figure(figsize=(8, 6))
plt.hist(df['Total'], bins=10, color='green', edgecolor='black')

# Add titles and labels
plt.title('Distribution of Total Deaths Across All Years')
plt.xlabel('Total Deaths')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
